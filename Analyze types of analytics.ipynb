```python
import pandas as pd
from tqdm import tqdm
import os
from nltk import sent_tokenize
```

```python
filepath = 'Statements/no_stop-phrases/2022/20221102.txt'
```

```python
def extract_sentences(doc):
    return sent_tokenize(doc)

dfs = []
with open(filepath, 'r', encoding='utf-8-sig') as f:
    document_date_yyyymmdd = os.path.basename(filepath).replace('.txt', '')
    meeting_date = f.readline().replace('MEETING_DATE: ','').replace('\n', '')
    doc = ''.join(f.readlines())

    sentences = extract_sentences(doc)

    one_doc_sentences_df = pd.DataFrame(sentences, columns=['sentence'])
    one_doc_sentences_df['document_date'] = document_date_yyyymmdd
    one_doc_sentences_df['meeting_date'] = meeting_date
    one_doc_sentences_df['sentence_index'] = one_doc_sentences_df.index

one_doc_sentences_df
```

### Types of analytics
* Descriptive: What happened?
* Diagnostic: Why did it happen?
* Predictive: What will happen?
* Prescriptive: How can we make it happen?

```python
analytic_type_dictionary = {'1':'Descriptive (1)', '2':'Diagnostic (2)', '3':'Predictive (3)', '4':'Prescriptive (4)', '0':'Voting'}
analytic_types, factors, results = [], [], []
for sentence in one_doc_sentences_df.sentence.values:
    print('\n\n', sentence)
    analytic_type = input()
    factor, result = '', ''
    if analytic_type in ('2','3'):
        factor = input('factor:')
        result = input('result:')
    analytic_types.append(analytic_type_dictionary[analytic_type])
    factors.append(factor)
    results.append(result)
    
one_doc_sentences_df['analytic_type'] = analytic_types
one_doc_sentences_df['factor'] = factors
one_doc_sentences_df['result'] = results
```

```python
one_doc_sentences_df['analytic_type'] = analytic_types
one_doc_sentences_df['factor'] = factors
one_doc_sentences_df['result'] = results

for _, row in one_doc_sentences_df.iterrows():
    print('{}: {}\n-> {}\n{}\n'.format(row['analytic_type'], row['factor'], row['result'], row['sentence']))
```
